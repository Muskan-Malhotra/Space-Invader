```python
import pygame

pygame.init()
pygame.font.init()

White = (255,255,255)
LIGHT = (255,255,0)
BLACK = ( 0, 0, 0)

font = pygame.font.SysFont('Comic Sans MS', 38)
myfont = pygame.font.SysFont('Comic Sans MS', 50)
Font = pygame.font.SysFont("Comic Sans MS", 55)
myfont_text = pygame.font.SysFont('Comic Sans MS', 35)

windw = pygame.display.set_mode((1500,1000))
pygame.display.set_caption('SPACE EXPLOSION')
clock=pygame.time.Clock()
pygame.mixer.music.load('sound.mp3')
pygame.mixer.music.play(-1,0.0)

done = False
level = 0
sprite1 = 5
sprite2 = 10
sprite3 = 15

backgrnd = pygame.image.load("space.jpg").convert()

page = 'mainscreen'

def exitpage():
    global done, page

    windw.blit(backgrnd, (0,0))
    mouse = pygame.mouse.get_pos()
    click = pygame.mouse.get_pressed()
    
    rect = pygame.image.load("rect0.jpg")
    rect = pygame.transform.scale(rect, (780,380))
    windw.blit(rect, (350,125))

    #pygame.draw.rect(windw, WHITE, (450, 300, 600, 300))

    textsurface = myfont.render('Do you really want to exit?', False, White)
    windw.blit(textsurface, (425,185))
    
    bot2=pygame.image.load("bot2.jpg")
    bot2 = pygame.transform.scale(bot2, (159, 65))
    windw.blit(bot2, (500, 348))
    
    if mouse[0] < 659 and mouse[0] > 500 and mouse[1] < 413 and mouse[1] > 348:
        if click[0] == 1:
            image1 = pygame.transform.scale(bot2, (130, 58))
            windw.blit(image1,(500,348))
            page = 'mainscreen'
        else:
            image1 = pygame.transform.scale(bot2, (160, 70))
            windw.blit(image1,(500,348))
    textsurface = myfont.render('Cancel', False, White)
    windw.blit(textsurface, (506,345))

    
    bot2=pygame.image.load("bot2.jpg")
    bot2 = pygame.transform.scale(bot2, (159, 65))
    windw.blit(bot2, (780, 348))
    
    if mouse[0] < 929 and mouse[0] > 780 and mouse[1] < 413 and mouse[1] > 348:
        if click[0] == 1:
            image1 = pygame.transform.scale(bot2, (159, 64))
            windw.blit(image1,(780,348))
            done = True
        else:
            image1 = pygame.transform.scale(bot2, (160, 70))
            windw.blit(image1,(780,348))
    textsurface = myfont.render('Exit', False, White)
    windw.blit(textsurface, (815,345))


def howtoplay():
    global page, sprite1, sprite2, sprite3

    mouse = pygame.mouse.get_pos()
    click = pygame.mouse.get_pressed()
    windw.blit(backgrnd, (0,0))

    textsurface = Font.render('Controls', False, White)
    windw.blit(textsurface, (660, 50))

    ctrls = pygame.image.load("keyboard.png").convert()
    windw.blit(ctrls, (450,150))

    pygame.draw.arc(windw, BLACK, (890,310,70,120), 180*3.14/180, 270*3.14/180, 3)
    pygame.draw.arc(windw, BLACK, (950,325,180,90), 180*3.14/180, 270*3.14/180, 3)
    pygame.draw.arc(windw, BLACK, (580,308,90,120), 270*3.14/180, 360*3.14/180, 3)
    textsurface = Font.render('Scoring', False, White)
    windw.blit(textsurface, (660, 465))

    for i in range(1,4):
        img = pygame.image.load('sprite'+str(i)+'.jpeg')
        img = pygame.transform.scale(img, (70,70))
        windw.blit(img, (600, 480+i*80))

    textsurface = font.render(str(sprite1)+'   points', False, LIGHT)
    windw.blit(textsurface, (720, 475+1*90))

    textsurface = font.render(str(sprite2)+'   points', False, LIGHT)
    windw.blit(textsurface, (720, 465+2*90))

    textsurface = font.render(str(sprite3)+'   points', False, LIGHT)
    windw.blit(textsurface, (720, 465+3*90))
    
    bot4=pygame.image.load("bot2.jpg")
    bot4 = pygame.transform.scale(bot4, (100, 55))
    windw.blit(bot4, (1300, 732))

    if mouse[0] < 1400 and mouse[0] > 1300 and mouse[1] < 787 and mouse[1] > 732:
        if click[0]:
            image1 = pygame.transform.scale(bot4, (100, 54))
            windw.blit(image1,(1300, 732))
            page = 'mainscreen'
        else:
            image1 = pygame.transform.scale(bot4, (120, 60))
            windw.blit(image1,(1300, 732))

    textsurface = myfont_text.render('Back', False, White)
    windw.blit(textsurface, (1310, 735))
    
    textsurface = myfont_text.render('SHOOT', False, LIGHT)
    windw.blit(textsurface, (490,400))

    textsurface = myfont_text.render('MOVE LEFT', False, LIGHT)
    windw.blit(textsurface, (830,425))

    textsurface = myfont_text.render('MOVE RIGHT', False, LIGHT)
    windw.blit(textsurface, (1040,385))

def newgame():
    global page, level

    mouse = pygame.mouse.get_pos()
    click = pygame.mouse.get_pressed()
    windw.blit(backgrnd, (0,0))
    
    rec = pygame.image.load("rect0.jpg")
    rec = pygame.transform.scale(rec, (650,755))
    windw.blit(rec, (650,70))
    
    bot3=pygame.image.load("bot2.jpg")
    bot3 = pygame.transform.scale(bot3, (350, 75))
    windw.blit(bot3, (740, 300))

    textsurface = Font.render('Difficulty Levels', False, White)
    windw.blit(textsurface, (775,150))

    if mouse[0] < 1100 and mouse[0] > 750 and mouse[1] < 375 and mouse[1] > 300:
        if click[0]:
            image1 = pygame.transform.scale(bot3, (355, 76))
            windw.blit(image1,(740,300))
            level = 1
        else:
            image1 = pygame.transform.scale(bot3, (365, 80))
            windw.blit(image1,(740,300))

    textsurface = myfont.render('Let me try', False, White)
    windw.blit(textsurface, (790, 300))
    
    bot3 = pygame.transform.scale(bot3, (350, 75))
    windw.blit(bot3, (800, 425))

    if mouse[0] < 1160 and mouse[0] > 810 and mouse[1] < 490 and mouse[1] > 425:
        if click[0]:
            image1 = pygame.transform.scale(bot3, (355, 76))
            windw.blit(image1,(800,425))
            level = 2
        else:
            image1 = pygame.transform.scale(bot3, (365, 80))
            windw.blit(image1,(800,425))

    textsurface = myfont.render('Bring them on', False, White)
    windw.blit(textsurface, (815, 427))
    
    bot3 = pygame.transform.scale(bot3, (350, 75))
    windw.blit(bot3, (870, 550))

    if mouse[0] < 1220 and mouse[0] > 870 and mouse[1] < 625 and mouse[1] > 550:
        if click[0]:
            image1 = pygame.transform.scale(bot3, (355, 76))
            windw.blit(image1,(870,550))
            level = 3
        else:
            image1 = pygame.transform.scale(bot3, (365, 80))
            windw.blit(image1,(870,550))
            
    textsurface = myfont.render('Begin the war', False, White)
    windw.blit(textsurface, (885, 553))
    
    bot3 = pygame.transform.scale(bot3, (150, 65))
    windw.blit(bot3, (950, 670))

    if mouse[0] < 1100 and mouse[0] > 950 and mouse[1] < 735 and mouse[1] > 670:
        if click[0]:
            image1 = pygame.transform.scale(bot3, (155, 66))
            windw.blit(image1,(950,670))
            page = 'mainscreen'
        else:
            image1 = pygame.transform.scale(bot3, (160, 70))
            windw.blit(image1,(950,670))

    textsurface = myfont.render('Back', False, White)
    windw.blit(textsurface, (970, 668))

def mainscreen():
    global done, page

    windw.blit(backgrnd, [0,0])
    title = pygame.image.load("grnd.jpg")
    title = pygame.transform.scale(title, (800,150))

    windw.blit(title, (360, 30))

    bot1=pygame.image.load("bot1.jpg")
    bot1 = pygame.transform.scale(bot1, (230, 65))

    windw.blit(bot1, (1060, 410))
    windw.blit(bot1, (1060, 485))
    windw.blit(bot1, (1060, 560))
    windw.blit(bot1, (1060, 635))
    windw.blit(bot1, (1060, 711))
    
    mouse = pygame.mouse.get_pos()
    click = pygame.mouse.get_pressed()
    if mouse[0] < 1290 and mouse[0]> 1060 and mouse[1] < 475 and mouse[1] > 410:
        if click[0] == 1:
            image = pygame.transform.scale(bot1, (230, 63))
            windw.blit(image,(1060,410))
            page = 'newgame'
        else:
            image = pygame.transform.scale(bot1, (240, 70))
            windw.blit(image,(1060,410))
            
        
    text = font.render('New Game', False, White)
    windw.blit(text,(1086,415))
    
    if mouse[0] < 1290 and mouse[0] > 1060 and mouse[1] < 550 and mouse[1] > 485:
        if click[0] == 1:
            image1 = pygame.transform.scale(bot1, (230, 63))
            windw.blit(image1,(1060,485))
        else:
            image1 = pygame.transform.scale(bot1, (240, 70))
            windw.blit(image1,(1060,485))
            
    text = font.render('Resume', False, White)
    windw.blit(text,(1108,490))
    
    if mouse[0] < 1290 and mouse[0] > 1060 and mouse[1] < 625 and mouse[1] > 560:
        if click[0] == 1:
            image2 = pygame.transform.scale(bot1, (230, 63))
            windw.blit(image2,(1060,560))
        else:
            image2 = pygame.transform.scale(bot1, (240, 70))
            windw.blit(image2,(1060,560))
            
    text = font.render('High Score', False, White)
    windw.blit(text,(1078,565))
    
    if mouse[0] < 1290 and mouse[0] > 1060 and mouse[1] < 700 and mouse[1] > 635:
        if click[0] == 1:
            image2 = pygame.transform.scale(bot1, (230, 63))
            windw.blit(image2,(1060,635))
            page = 'howtoplay'
        else:
            image2 = pygame.transform.scale(bot1, (240, 70))
            windw.blit(image2,(1060,635))
            
    text = font.render('How to Play', False, White)
    windw.blit(text,(1075,638))
    
    if mouse[0] < 1290 and mouse[0] > 1060 and mouse[1] < 776 and mouse[1] > 711:
        if click[0] == 1:
            image2 = pygame.transform.scale(bot1, (230, 63))
            windw.blit(image2,(1060,711))
            page = 'exitpage'
        else:
            image2 = pygame.transform.scale(bot1, (240, 70))
            windw.blit(image2,(1060,711))
            
    text = font.render('Exit', False, White)
    windw.blit(text,(1138,718))


#-----------------------------------------------------------------------------------

while done == False:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True


    if page == 'mainscreen':
        mainscreen()
    elif page == 'newgame':
        newgame()
    elif page == 'howtoplay':
        howtoplay()
    elif page == 'exitpage':
        exitpage()

    
    pygame.display.update()




    clock.tick(20) 

pygame.quit()
```
